# Feature Engineering

```python
color_labels = pd.read_csv('./petfinder_data/color_labels.csv')
```

```python
new_row = {'ColorID' : 0, 'ColorName' : 'None'}
color_labels = color_labels.append(new_row, ignore_index=True)
color_labels
```

```python
color_labels.to_csv('./petfinder_data/color_labels.csv', index = False)
```

```python
df.head(1)
```

```python
df = pd.merge(df, color_labels, left_on = 'Color1', right_on = 'ColorID')
df = df.drop(columns = ['ColorID'])
df['Colors'] = df['ColorName']
df = df.drop(columns = ['ColorName'])

df = pd.merge(df, color_labels, left_on = 'Color2', right_on = 'ColorID')
df['Colors'] = df['ColorName'] + ', ' + df['Colors']
df = df.drop(columns = ['ColorID', 'ColorName'])

df = pd.merge(df, color_labels, left_on = 'Color3', right_on = 'ColorID')
df['Colors'] = df['ColorName'] + ', ' + df['Colors']
df.drop(columns = ['ColorName', 'ColorID'], inplace=True)

color_dummies = df['Colors'].str.get_dummies(", ")

df = pd.merge(df, color_dummies, left_index =True, right_index=True)

df = df.drop(columns = ['None', 'Colors', 'Color1', 'Color2', 'Color3'])

df.head()
```

## - Count Vectorizing

```python
dog.Description.isnull().sum()
```

```python
# Dropping the two rows without descriptions prior to count vectorizing
dog = dog[dog['Description'].notna()]
```

```python
dog.Description.isnull().sum()
```

```python
dog_list = ['home', 'dog', 'puppy', 'adoption', 'adopt', 'looking', 'interested', 'dogs', 'adopter', 'like', 'time', 'months', 'new', 'just', 'need']
```

```python
stop_words = text.ENGLISH_STOP_WORDS.union(dog_list)
```

```python
dog_cvec= CountVectorizer(stop_words=stop_words)
X = dog['Description']
dog_cvec.fit(X)
X_trans = dog_cvec.transform(X)
```

```python
X_trans = pd.DataFrame(X_trans.todense(), 
                 columns=dog_cvec.get_feature_names())

plt.figure(figsize = [13, 9])
plt.title('Most Common Words in Dog Descriptions')
plt.xlabel('Word Count')
plt.ylabel('Word')
X_trans.sum().sort_values(ascending = False).head(20).plot(kind='barh');
```

## Playing with stop words ##

```python
cat_list = ['cat', 'home', 'kitten', 'cats', 'adoption', 'interested', 'new', 'looking', 'adopt', 'like', 'contact', 'house', 'time', 'months', 'just']
```

```python
stop_words = text.ENGLISH_STOP_WORDS.union(cat_list)
```

```python
cat_cvec= CountVectorizer(stop_words=stop_words)
X = cat['Description']
cat_cvec.fit(X)
X_trans = cat_cvec.transform(X)
```

```python
X_trans = pd.DataFrame(X_trans.todense(), 
                 columns=cat_cvec.get_feature_names())

plt.figure(figsize = [13, 9])
plt.title('Most Common Words in Dog Descriptions')
plt.xlabel('Word Count')
plt.ylabel('Word')
X_trans.sum().sort_values(ascending = False).head(20).plot(kind='barh');
```
