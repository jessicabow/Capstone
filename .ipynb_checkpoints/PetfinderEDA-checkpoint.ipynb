```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction import text
```

```python
dogs = pd.read_csv('./petfinder_data/dogs_eda.csv')
```

```python
dogs.columns
```

```python
dogs['year'] = pd.DatetimeIndex(dogs['published_at']).year
dogs.head(1)
```

```python
dogs['year_adopted'] = pd.DatetimeIndex(dogs['status_changed_at']).year
dogs.head(1)
```

```python
dogs['month_adopted'] = pd.DatetimeIndex(dogs['status_changed_at']).month
dogs.head(1)
```

```python
dogs.rename(columns={'days_on_petfinder': 'timely_adoption'})
```

```python
# Converting columns to datetime

dogs['published_at']= pd.to_datetime(dogs['published_at'])
dogs['status_changed_at']= pd.to_datetime(dogs['status_changed_at'])
```

```python
dogs['days_on_petfinder'] = dogs['status_changed_at'] - dogs['published_at']
dogs['days_on_petfinder'] = dogs['days_on_petfinder'] / np.timedelta64(1, 'D')
dogs['days_on_petfinder'] = dogs['days_on_petfinder'].round()
```

```python
dogs.drop(columns = ['id', 'url', 'type', 'species', 'tags', 'organization_animal_id', 'photos', 'videos', 'status', 'breeds.unknown',
          'attributes.declawed', 'primary_photo_cropped.small', 'primary_photo_cropped.medium', 'primary_photo_cropped.large', 
          'primary_photo_cropped.full', 'contact.email', 'contact.phone', 'contact.address.address1', 'contact.address.address2',
          'contact.address.country', 'animal_id', 'animal_type', 'organization_id.1', 'primary_photo_cropped'], inplace=True)
```

```python
dogs['good_with_kids'] = dogs['good_with_kids'].fillna(2)
dogs['good_with_dogs'] = dogs['good_with_dogs'].fillna(2)
dogs['good_with_cats'] = dogs['good_with_cats'].fillna(2)

dogs['fixed'] = (dogs['fixed'] * 1).astype(int)
dogs['house_trained'] = (dogs['house_trained'] * 1).astype(int)
dogs['shots_current'] = (dogs['shots_current'] * 1).astype(int)
dogs['good_with_kids'] = (dogs['good_with_kids'] * 1).astype(int)
dogs['good_with_dogs'] = (dogs['good_with_dogs'] * 1).astype(int)
dogs['good_with_cats'] = (dogs['good_with_cats'] * 1).astype(int)
dogs['special_needs'] = (dogs['special_needs'] * 1).astype(int)
```

```python
dogs['coat'].fillna('Not specified', inplace=True)
dogs['colors.primary'].fillna('Not specified', inplace=True)
dogs['breeds.secondary'].fillna('Not specified', inplace=True)
```

```python
dogs['breeds.mixed'].value_counts()
```

```python
dogs.isnull().sum()
```

```python
dogs.drop(columns=['name', 'description', 'status_changed_at', 'published_at', 'colors.secondary', 'colors.tertiary', 
                   'zipcode'], inplace=True)
```
